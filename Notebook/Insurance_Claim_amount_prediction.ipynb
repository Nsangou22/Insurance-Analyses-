```python
dataset = pd.read_csv(r"C:\Users\nabde\OneDrive - aivancity\Desktop\Projet Github\Insurance Project\Insurance Policies .csv")
columns_to_keep = [
    'birthdate',           # for age
    'marital_status',
    'car_use',
    'gender',
    'kids_driving',
    'parent',
    'education',
    'car_make',
    'car_year',
    'claim_freq',
    'coverage_zone',
    'household_income',
    'claim_amt'            # target variable
]

df = dataset[columns_to_keep]


df
```

```python


# Example: assume your dataframe is called df and birthdate is a string (like '08/09/1962')
# Convert birthdate to datetime
df['birthdate'] = pd.to_datetime(df['birthdate'], errors='coerce')

# Extract birth year
df['birth_year'] = df['birthdate'].dt.year

# Get current year
current_year = datetime.now().year

# Calculate age
df['age'] = current_year - df['birth_year']

df
```

## Encoding Categorical Variable

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

categorical_cols = x.select_dtypes(include=['object']).columns.tolist()

ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(drop='first', sparse_output=False), categorical_cols)
    ],
    remainder='passthrough'  # keep other columns as is
)

X_encoded = ct.fit_transform(x)


X_encoded
```

## Splitting Into training and testing 

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test =  train_test_split( X_encoded, y , test_size=0.20, random_state=42)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train= sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

## Applying PCA 

```python
from sklearn.decomposition import PCA
pca = PCA()  
pca.fit_transform(x_train)

pca.n_components_
```

```python
eig = pd.DataFrame(
    {
        "Dimension" : ["Dim" + str(x+1) for x in range(90)],
        "Explained Variance" : pca.explained_variance_,
        "% explained variance" : np.round(pca.explained_variance_ratio_*100),
        "% cumulative expl. variance" : np.round(np.cumsum(pca.explained_variance_ratio_)*100)
    }
)
eig.head(80)
```

```python
pca = PCA(n_components=76) 
X_train_pca = pca.fit_transform(x_train)
X_test_pca = pca.transform(x_test)
```

## Training the Model

```python
## Random Forest

from sklearn.ensemble import RandomForestRegressor
Regressor = RandomForestRegressor()
Regressor.fit(x_train,y_train)
```

```python
import xgboost as xgb


xg_reg = xgb.XGBRegressor(random_state=42)
xg_reg.fit(x_train, y_train)
```

## Evaluating the Model

```python
y_pred = xg_reg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt

# Assuming y_test and y_pred are defined
mse = mean_squared_error(y_test, y_pred)
rmse = sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R² Score: {r2:.4f}")
```

## UI Interface

```python
import sys
from PyQt5.QtWidgets import (QApplication, QMainWindow, QWidget, QVBoxLayout, QHBoxLayout, 
                            QLabel, QPushButton, QComboBox, QGroupBox, QFrame,
                            QFormLayout, QMessageBox, QDoubleSpinBox, QSpinBox, QSpacerItem,
                            QSizePolicy)
from PyQt5.QtCore import Qt, QSize
from PyQt5.QtGui import QFont, QPalette, QColor, QIcon
import pandas as pd

class ClaimPredictorApp(QMainWindow):
    def __init__(self, model, ct, sc=None):
        super().__init__()
        self.model = model
        self.ct = ct
        self.sc = sc
        
        self.setWindowTitle("Insurance Claim Predictor Pro")
        self.setWindowIcon(QIcon("C:\Users\nabde\OneDrive - aivancity\Desktop\Projet Github\Insurance Project\insurance.png")) 
        self.setMinimumSize(700, 700)
        
       
        self.setStyleSheet("""
            QMainWindow {
                background-color: #f5f7fa;
            }
        """)
        
        self.init_ui()
        
    def init_ui(self):
        
        central_widget = QWidget()
        central_widget.setObjectName("centralWidget")
        self.setCentralWidget(central_widget)
        
        
        main_layout = QVBoxLayout(central_widget)
        main_layout.setContentsMargins(20, 20, 20, 20)
        main_layout.setSpacing(15)
        
        
        header = QLabel("CLAIM AMOUNT PREDICTOR")
        header.setAlignment(Qt.AlignCenter)
        header.setStyleSheet("""
            QLabel {
                font-size: 24px;
                font-weight: bold;
                color: #2c3e50;
                padding: 10px;
                border-bottom: 2px solid #3498db;
            }
        """)
        main_layout.addWidget(header)
        
        
        main_layout.addSpacing(10)
        
        
        form_container = QFrame()
        form_container.setFrameShape(QFrame.StyledPanel)
        form_container.setStyleSheet("""
            QFrame {
                background-color: white;
                border-radius: 8px;
                padding: 15px;
                border: 1px solid #dfe6e9;
            }
        """)
        
        form_layout = QFormLayout()
        form_layout.setVerticalSpacing(15)
        form_layout.setHorizontalSpacing(20)
        form_layout.setContentsMargins(15, 15, 15, 15)
        
        
        self.create_input_fields(form_layout)
        
        form_container.setLayout(form_layout)
        main_layout.addWidget(form_container)
        
        
        predict_btn = QPushButton("PREDICT CLAIM AMOUNT")
        predict_btn.setCursor(Qt.PointingHandCursor)
        predict_btn.setFixedHeight(45)
        predict_btn.clicked.connect(self.predict_claim)
        predict_btn.setStyleSheet("""
            QPushButton {
                background-color: #3498db;
                color: white;
                font-weight: bold;
                border: none;
                border-radius: 4px;
                padding: 10px;
                font-size: 16px;
            }
            QPushButton:hover {
                background-color: #2980b9;
            }
            QPushButton:pressed {
                background-color: #1c6ea4;
            }
        """)
        main_layout.addWidget(predict_btn)
        
        
        self.result_display = QLabel("Prediction will appear here")
        self.result_display.setAlignment(Qt.AlignCenter)
        self.result_display.setStyleSheet("""
            QLabel {
                font-size: 20px;
                font-weight: bold;
                color: #27ae60;
                background-color: white;
                border-radius: 8px;
                padding: 15px;
                border: 1px solid #dfe6e9;
                min-height: 30px;
            }
        """)
        main_layout.addWidget(self.result_display)
        
        
        footer = QLabel("© 2023 Insurance Analytics Pro")
        footer.setAlignment(Qt.AlignCenter)
        footer.setStyleSheet("""
            QLabel {
                color: #7f8c8d;
                font-size: 12px;
                padding-top: 15px;
                border-top: 1px solid #dfe6e9;
            }
        """)
        main_layout.addWidget(footer)
        
        
        self.setTabOrder(self.car_use_combo, self.kids_driving_spin)
        self.setTabOrder(self.kids_driving_spin, self.education_combo)
        
        
    def create_input_fields(self, form_layout):
        
        label_style = """
            QLabel {
                font-weight: bold;
                color: #2c3e50;
                font-size: 14px;
            }
        """
        
        combo_style = """
            QComboBox {
                padding: 8px;
                border: 1px solid #bdc3c7;
                border-radius: 4px;
                min-width: 150px;
            }
            QComboBox:hover {
                border: 1px solid #3498db;
            }
        """
        
        spinbox_style = """
            QSpinBox, QDoubleSpinBox {
                padding: 8px;
                border: 1px solid #bdc3c7;
                border-radius: 4px;
                min-width: 150px;
            }
            QSpinBox:hover, QDoubleSpinBox:hover {
                border: 1px solid #3498db;
            }
        """
        
        # Car Use
        self.car_use_combo = QComboBox()
        self.car_use_combo.addItems(["Private", "Commercial"])
        self.car_use_combo.setStyleSheet(combo_style)
        form_layout.addRow(QLabel("Car Use:"), self.car_use_combo)
        
        # Kids Driving
        self.kids_driving_spin = QSpinBox()
        self.kids_driving_spin.setRange(0, 10)
        self.kids_driving_spin.setStyleSheet(spinbox_style)
        form_layout.addRow(QLabel("Kids Driving:"), self.kids_driving_spin)
        
        # Education Level
        self.education_combo = QComboBox()
        self.education_combo.addItems(["High School", "Bachelors", "Masters", "PhD"])
        self.education_combo.setStyleSheet(combo_style)
        form_layout.addRow(QLabel("Education:"), self.education_combo)
        
        # Car Make
        self.car_make_combo = QComboBox()
        self.car_make_combo.addItems(["Acura", "Corbin", "Nissan", "Ford", "Toyota", "Honda", "BMW", "Mercedes"])
        self.car_make_combo.setStyleSheet(combo_style)
        form_layout.addRow(QLabel("Car Make:"), self.car_make_combo)
        
        # Car Year
        self.car_year_spin = QSpinBox()
        self.car_year_spin.setRange(1900, 2023)
        self.car_year_spin.setValue(2010)
        self.car_year_spin.setStyleSheet(spinbox_style)
        form_layout.addRow(QLabel("Car Year:"), self.car_year_spin)
        
        # Claim Frequency
        self.claim_freq_spin = QSpinBox()
        self.claim_freq_spin.setRange(0, 10)
        self.claim_freq_spin.setStyleSheet(spinbox_style)
        form_layout.addRow(QLabel("Claim Frequency:"), self.claim_freq_spin)
        
        # Coverage Zone
        self.coverage_zone_combo = QComboBox()
        self.coverage_zone_combo.addItems(["Highly Urban", "Urban", "Rural"])
        self.coverage_zone_combo.setStyleSheet(combo_style)
        form_layout.addRow(QLabel("Coverage Zone:"), self.coverage_zone_combo)
        
        # Household Income
        self.household_income_spin = QDoubleSpinBox()
        self.household_income_spin.setRange(0, 1000000)
        self.household_income_spin.setValue(50000)
        self.household_income_spin.setPrefix("$ ")
        self.household_income_spin.setSingleStep(1000)
        self.household_income_spin.setStyleSheet(spinbox_style)
        form_layout.addRow(QLabel("Household Income:"), self.household_income_spin)
        
        # Age
        self.age_spin = QSpinBox()
        self.age_spin.setRange(18, 100)
        self.age_spin.setValue(30)
        self.age_spin.setStyleSheet(spinbox_style)
        form_layout.addRow(QLabel("Policyholder Age:"), self.age_spin)
        
        # Apply label styles
        for i in range(form_layout.rowCount()):
            label = form_layout.itemAt(i, QFormLayout.LabelRole).widget()
            label.setStyleSheet(label_style)
    
    def predict_claim(self):
        try:
            input_data = {
                'car_use': self.car_use_combo.currentText(),
                'kids_driving': self.kids_driving_spin.value(),
                'education': self.education_combo.currentText(),
                'car_make': self.car_make_combo.currentText(),
                'car_year': self.car_year_spin.value(),
                'claim_freq': self.claim_freq_spin.value(),
                'coverage_zone': self.coverage_zone_combo.currentText(),
                'household_income': self.household_income_spin.value(),
                'age': self.age_spin.value()
            }
            
            input_df = pd.DataFrame([input_data])
            X_encoded = self.ct.transform(input_df)
            X_scaled = self.sc.transform(X_encoded) if self.sc else X_encoded
            prediction = self.model.predict(X_scaled)
            
            # Format the prediction nicely
            formatted_pred = f"${prediction[0]:,.2f}"
            self.result_display.setText(f"Predicted Claim Amount: {formatted_pred}")
            
            # Add visual feedback
            self.result_display.setStyleSheet("""
                QLabel {
                    font-size: 20px;
                    font-weight: bold;
                    color: #27ae60;
                    background-color: white;
                    border-radius: 8px;
                    padding: 15px;
                    border: 1px solid #2ecc71;
                    min-height: 30px;
                }
            """)
            
        except Exception as e:
            self.result_display.setText("Error in prediction")
            self.result_display.setStyleSheet("""
                QLabel {
                    font-size: 20px;
                    font-weight: bold;
                    color: #e74c3c;
                    background-color: white;
                    border-radius: 8px;
                    padding: 15px;
                    border: 1px solid #e74c3c;
                    min-height: 30px;
                }
            """)
            QMessageBox.critical(self, "Error", f"Prediction failed:\n{str(e)}")

def launch_app(model, ct, sc=None):
    app = QApplication(sys.argv)
    
    # Set application font
    font = QFont()
    font.setFamily("Segoe UI")  # Or any modern font
    font.setPointSize(10)
    app.setFont(font)
    
    # Set palette for consistent colors
    palette = QPalette()
    palette.setColor(QPalette.Window, QColor(245, 247, 250))
    palette.setColor(QPalette.WindowText, QColor(44, 62, 80))
    palette.setColor(QPalette.Base, QColor(255, 255, 255))
    palette.setColor(QPalette.AlternateBase, QColor(233, 236, 239))
    palette.setColor(QPalette.ToolTipBase, QColor(255, 255, 255))
    palette.setColor(QPalette.ToolTipText, QColor(44, 62, 80))
    palette.setColor(QPalette.Text, QColor(44, 62, 80))
    palette.setColor(QPalette.Button, QColor(52, 152, 219))
    palette.setColor(QPalette.ButtonText, QColor(255, 255, 255))
    palette.setColor(QPalette.BrightText, QColor(255, 0, 0))
    palette.setColor(QPalette.Highlight, QColor(52, 152, 219))
    palette.setColor(QPalette.HighlightedText, QColor(255, 255, 255))
    app.setPalette(palette)
    
    window = ClaimPredictorApp(model, ct, sc)
    window.show()
    app.exec_()
```
